```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

# Load some sample data

```python
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from keras.utils import to_categorical
```

```python
digits = load_digits()
X, y = digits.data, digits.target
y_cat = to_categorical(y, 10)
X_train, X_test, y_train, y_test = train_test_split(X, y_cat,
                                                    test_size=0.3)
```

```python
for i in range(8):
    plt.subplot(1,8,i+1)
    plt.imshow(X.reshape(-1, 8, 8)[i], cmap='gray')
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import to_categorical
import keras.backend as K
from keras.callbacks import EarlyStopping
from keras.layers import Dropout
```

```python
learning_keep_prob_l1 = 0.2
learning_keep_prob_l2 = 0.4

model = Sequential()
model.add(Dropout(rate = (1-learning_keep_prob_l1), input_shape=X_train.shape[1:]))
# first fully connected layer
model.add(Dense(512, kernel_initializer='normal',
                kernel_regularizer='l2', activation='sigmoid'))
model.add(Dropout(rate=(1-learning_keep_prob_l2)))
model.add(Dense(10, activation='softmax'))

model.compile('sgd',
              'categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, validation_data=(X_test, y_test), batch_size=32, epochs=300)
```

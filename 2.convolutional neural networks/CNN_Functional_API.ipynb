# Import libraries

```python
import numpy as np
from keras.models import Input
from keras.layers import Conv2D
from keras.layers import AvgPool2D
from keras.layers import Activation
from keras.layers import Flatten
from keras.layers import Dense
from keras.models import Model
from keras.callbacks import ModelCheckpoint
from keras.utils.np_utils import to_categorical
import keras.backend as kb
```

# import data

```python
from keras.datasets import mnist
(X_train, y_train), (X_test, y_test) = mnist.load_data('/tmp/mnist.npz')
```

# Assuming the data in Gray scale 
- thus 1 channel with 256 color range
- pixels width & hieght = 28

```python
X_train = X_train.reshape(-1, 28,28, 1)
X_test = X_test.reshape(-1, 28,28, 1)
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
```

```python
input_layer = Input(shape=(28,28,1))
conv_layer = Conv2D(filters= 32, kernel_size=(3,3)) (input_layer)
avg_pool_layer = AvgPool2D(pool_size=(2,2)) (conv_layer)
actiation_layer = Activation(activation='relu') (avg_pool_layer)
flatten_layer = Flatten() (actiation_layer)
neural_network_first_layer = Dense(units=256, activation='relu') (flatten_layer)
neural_network_second_layer = Dense(units=128, activation='relu') (neural_network_first_layer)
neural_network_third_layer = Dense(units=64, activation='relu') (neural_network_second_layer)
output_layer = Dense(units=10, activation='softmax') (neural_network_third_layer)

model = Model(inputs = input_layer, outputs = output_layer)
model.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics = ['accuracy'])

checkpoint = ModelCheckpoint(filepath='/content/gdrive/My Drive/deep-learning-vinsmon/mnist_functional_api_model.hdf5', save_best_only=True)
```

```python
model.summary()
```

```python
model.fit(X_train, y_train, epochs=15, batch_size=32, callbacks=[checkpoint], verbose=1, validation_data=(X_test, y_test))
```
